```python
import pandas as pd

# Load Dataset
df = pd.read_csv('Dataset/Superstore.csv', encoding='latin1')

# Preview data
df.head()
```

```python
#Data Inspection
#check datatypes

#check column types and non-null counts
df.info()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
#Summary Statistics for numerical columns
df.describe()
```

```python
#Unique Values in 'Category' and 'Department'
print("Categories:",df['Category'].unique())
print("Departments:",df['Department'].unique())

#unique values in products
print('Items:',df ['Item'].unique()[:10])
```

```python
df.columns=df.columns.str.strip()
```

```python
print(df.columns.tolist())
```

```python
import matplotlib.pyplot as plt

df.groupby('State')['Sales'].sum().sort_values(ascending=False).head(10).plot(kind='bar', figsize=(10,6))
plt.title("Top 10 States by Sales")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
df.groupby('Category')['Profit'].sum().plot(kind='bar', color='skyblue')
plt.title("Profit by Category")
plt.ylabel("Total Profit")
plt.xticks(rotation=0)
plt.show()
```
